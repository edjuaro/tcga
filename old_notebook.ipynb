## Making GCT file for differential expression analysis

```python
print(normal_df.shape, cancer_df.shape)
print("Desired shape is: (",len(normal_df),",",normal_df.shape[1]+cancer_df.shape[1],")")
```

```python
df = normal_df
df = df.join(cancer_df)

df.shape
```

##### Keep only top 40% (~10k) most variable genes

```python
variances = np.nanvar(df.as_matrix().astype(float), axis=1)
```

```python
# variance_dict = {}
# for index, row in df.iterrows():
#     variance_dict[index] = row.var()

# print(sorted(variances))
# print(len(variances))
# how_many = 60
min_var = np.percentile(variances,50)
# print(min_var)
how_many=(variances > min_var).sum()
print(how_many)
```

```python
df['variance'] = variances
df.sort_values(by='variance', ascending=False, inplace=True)
df.drop('variance',1, inplace=True)
df
```

# Renaming Gene IDs

```python
genes = df.index
names = []
IDs = []
for gene in genes:
    g_list = gene.split('|')
    names.append(g_list[0])
    IDs.append('NCBI Gene ID: ' + g_list[1])
df.index = names
df.index.names = ['Name']
df['Description'] = IDs
description = df['Description']
df.drop(labels=['Description'], axis=1,inplace = True)
df.insert(0, 'Description', description)
```

# Removing genes with low variance

```python
filtered_df = df.head(how_many)
filtered_df
```

```python
print(filtered_df.shape)
```

## Enforcing no-duplicate-rows rule

```python
# duplicates = filtered_df.loc[filtered_df.index.duplicated()]
# for temp in np.unique(duplicates.index):
#     new_names = filtered_df.loc[temp].index.values+'-'+filtered_df.loc[temp, 'Description']
#     filtered_df.loc[temp].index = new_names.values
#     print(filtered_df.loc[temp].index)
duplicates = filtered_df.loc[filtered_df.index.duplicated()]
print(duplicates.shape)

filtered_df['to_drop'] = filtered_df.index
filtered_df.drop_duplicates(subset='to_drop', keep='first', inplace=True)
filtered_df.drop('to_drop', axis=1, inplace=True)

duplicates = filtered_df.loc[filtered_df.index.duplicated()]
print(duplicates.shape)

```

# Saving the GCT file

```python
cusca.df2gct(df=filtered_df, extra_columns=1, use_index=True, name='BRCA_workshop.gct', add_dummy_descriptions=False)
```
